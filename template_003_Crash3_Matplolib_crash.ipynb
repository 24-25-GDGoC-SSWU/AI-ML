### title, label 추가

````python
x = [5,8,10]
y = [12,16,6]

x2 = [6,9,11]
y2 = [6,15,7]

plt.plot(x,y)
plt.plot(x2,y2)
plt.title("Test Plot")
plt.xlabel("month")
plt.ylabel("price")
plt.show()
````

## OOP style

````python
np.random.randint(1,11,50) #이건 무조건 범위를 주게 되어있음 1~11 사이의 랜덤 난수를 50개 줘
````

````python
x = np.random.randint(low=1, high=11, size=50) # 11은 불포함 
y = x + np.random.randint(1, 5, size=x.size) # x.size은 50

data = x + y
data
````

### multiple subplots

**y = x + noise 인 (x, y) 쌍의 toy data 생성**

````python
fig,(ax1,ax2) = plt.subplots(1,2,figsize = (8,4)) # fig, 그림판을 선언해줌
# 그림판에 ax1,ax2 두 개의 도화지를 배치할거다 (ax1,ax2)
# plt.subplots(1,2) --> 행이 하나, 열이 두 개인 형태로 그림판을 두 개 배치하겠다
# 전체 그림 사이즈, figsize는 8,4 --> 가로 8, 세로 4로 만들겠다.(그림판의 사이즈)
ax1.scatter(x,y) #ax1에는 scatterplot을 그림, x가 얼마일때 y가 얼마이다~
ax1.scatter(x,y,c = "r" , marker = "o", edgecolor = "k")
ax1.set_title("SCATTER PLOT")
ax1.set_xlabel("X axis")
ax1.set_ylabel("Y axis") 

ax2.hist(data, bins = 20) # 두 번째 그림판에는 히스토그램으로 그래프 표현 
ax2.set_title("HISTOGRAM")
plt.show()
````

## Imshow

image data 처럼 행과 열을 가진 행렬 형태의 2차원 데이터는 imshow 로 표시

````python
from sklearn.datasets import load_digits
````

````python
x = load_digits() # load_digits 데이터 중에서 우리는 이미지 데이터를 쓰려고 함
# 파이썬의 딕셔너리 형태로 이뤄져있음 
img = x['images']
img
img.shape #1797개의 데이터의 이미지가 8 by 8 형태로 되어 있
````

````python
img = x['images'][0]
img.shape # 사진 딱 한 장만 시각화 해보자, 1797개의 이미지 중에서 한 개만
````

````python
plt.imshow(img) #imshow는 이미지를 보이게 하는 코드
````

````python
plt.imshow(img, cmap ='gray') # 흑백 ver
plt.xticks([]) # 빈리스트로 바꿔줌 --> 눈금이 사라짐 
plt.yticks([]) # 이러면 x,y축의 눈금 없어짐 
````
